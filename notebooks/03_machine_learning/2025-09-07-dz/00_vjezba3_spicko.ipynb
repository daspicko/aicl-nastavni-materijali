```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
diabetes_df = pd.read_csv("./diabetes_dataset.csv", low_memory=False)
```

```python
diabetes_df
```

```python
target=diabetes_df["Outcome"]
x_num = diabetes_df.select_dtypes(include=np.number)
```

```python
x_num
```

```python
imputer = SimpleImputer(strategy="median")
x_num_imputed = pd.DataFrame(imputer.fit_transform(x_num), columns=x_num.columns)
```

```python
x_num_imputed
```

```python
diabetes_df_cleaned = pd.concat([x_num_imputed, target], axis=1)
```

```python
diabetes_df
```

```python
y = diabetes_df_cleaned["Outcome"].astype(int)
y
```

```python
x = diabetes_df_cleaned.drop(columns=["Outcome"]).select_dtypes(include=np.number)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)
y_predictions = model.predict(x_test)
accuracy = accuracy_score(y_test, y_predictions)
```
