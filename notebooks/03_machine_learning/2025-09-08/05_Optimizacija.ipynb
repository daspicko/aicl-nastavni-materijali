```python
import numpy as np
import pandas as pd
```

### Slučajna vrijednost

```python
target = np.array([3, 5, 8, 10, 12, 15])
num_guess = 100
rezultati = []
for i in range(num_guess):
    guess = np.random.uniform(low=target.min(), high=target.max())
    #print("Guess", guess)
    total_distance = np.sum((target - guess) ** 2)
    #print("Total distance", total_distance)
    rezultati.append((total_distance, guess))

best = sorted(rezultati)[0][1]
print("Best", best)
```

### Slučajan korak

```python
target = np.array([3, 5, 8, 10, 12, 15])
num_steps = 1000
step_size = 0.05

best_guess = np.random.uniform(low=target.min(), high=target.max())
best_distance = np.sum((target - best_guess) ** 2)
for i in range(num_guess):
    new_guess = best_guess + (np.random.choice([+1, -1]) * step_size)
    new_distance = np.sum((target - new_guess) ** 2)
    if new_distance < best_distance:
        best_guess = new_guess
        best_distance = new_distance

print("Best guess", best_guess)
```

### Pametan korak

```python
target = np.array([3, 5, 8, 10, 12, 15])
num_steps = 100
step_size = 0.02
best_guess = np.random.uniform(low=target.min(), high=target.max())
best_distance = np.sum((target - best_guess) ** 2)
for i in range(num_guess):
    guesses = best_guess + (np.array([+1, -1]) * step_size)
    distances = np.sum((target[:,np.newaxis] - guesses) ** 2, axis=0)
    better_index = np.argmin(distances)

    if distances[better_index] > best_distance:
        break
    best_guess = guesses[better_index]
    best_distance = distances[better_index]

print("Best guess", best_guess)
```
