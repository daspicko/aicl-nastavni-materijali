```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'X': [1,3,6],
    'Y': [1,6,3],
    'Z': [10,5,1]
})
```

```python
data.index.name = 'Primjeri'
display(data)
display(data.cov())
```

```python
fix, ax = plt.subplots(1, 1, figsize=(4, 3))
data.plot(ax=ax)
ax.vlines([0,1,2,3], 0, 10, colors=".5")
ax.legend(loc='lower center', ncol=3)
plt.box(False)
ax.set_xticks([0,1,2,3])
ax.set_ylabel('values')
```
