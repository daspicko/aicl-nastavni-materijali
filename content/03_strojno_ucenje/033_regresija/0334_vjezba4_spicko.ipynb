Iz dataseta izbacite sve značajke koje nisu numeričke, a ako negdje nedostaje vrijednost umetnite medijan. 20% dataseta ostavite za testiranje, a pri dijeljenju dataset-a u funkciji train_test_split koristite `random_state=42`. Istrenirajte `k-NN regresiju` sa `3 susjeda` i ispišite koja je srednja kvadratna pogreška.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
```

```python
df = pd.read_csv("./housing.csv", low_memory=False)
df.head()
```

```python
target = df["median_house_value"]
x_num = df.select_dtypes(include=np.number)

imputer = SimpleImputer(strategy="median")
x_num_imputed = pd.DataFrame(imputer.fit_transform(x_num), columns=x_num.columns)

df_cleaned = pd.concat([x_num_imputed, target], axis=1)

y = df_cleaned["median_house_value"]
x = df_cleaned.drop(columns=["median_house_value"]).select_dtypes(include=np.number)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

knn = KNeighborsRegressor(n_neighbors=3)
model = knn.fit(x_train, y_train)
predictions = model.predict(x_test)
metrics.mean_squared_error(y_test, predictions)
```

```python
print("Srednja kvadratna pogreška", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
