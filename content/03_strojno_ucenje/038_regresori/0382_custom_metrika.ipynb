```python
import numpy as np
from sklearn import datasets, dummy, metrics, neighbors
from sklearn.model_selection import train_test_split, cross_val_score
import pandas as pd
```

```python
def rms_error(actual, predicted):
    mse = metrics.mean_squared_error(actual, predicted)
    return np.sqrt(mse)
```

```python
def negative_rmse_score(actual, predicted):
    return -rms_error(actual, predicted)
```

```python
def negative_rmse_scorer(mod, features, target):
    predictions = mod.predict(features)
    return negative_rmse_score(target, predictions)
```

```python
diabetes = datasets.load_diabetes()
knn = neighbors.KNeighborsRegressor(n_neighbors=3)
cross_val_score(knn, diabetes.data, diabetes.target, cv=5, scoring=negative_rmse_scorer)
```
