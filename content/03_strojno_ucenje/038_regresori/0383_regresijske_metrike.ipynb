# R^2

Koeficijent determinacije - koliki postatak varijance naš model objašnjava

## Zadana metrika za regresije

$$
SSE_{ours} = \sum
$$

$$
SSE_{mean} = \sum_i^n meanErrors_i^2 = \sum_i^n (meanPreds_i - actual_i)^2
$$

$$
R^2 = 1 - \frac{SSE_{ours}}{SSE_{mean}}
$$

MAE (Mean Absolute Error)

MSE (Mean Squared Error)

Median Absolute Error

```python
import numpy as np
```

```python
our_predictions = np.array([1, 2, 3])
mean_predictions = np.array([2, 2, 2])
actual = np.array([2, 3, 4])
```

```python
sse_ours = np.sum((our_predictions - actual)**2)
sse_mean = np.sum((mean_predictions - actual)**2)
r_squared = 1 - sse_ours / sse_mean
print(r_squared)
```
