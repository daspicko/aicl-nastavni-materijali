```python
from sklearn import datasets, dummy, metrics
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
diabetes = datasets.load_diabetes()
train_data, test_data, train_target, test_target = train_test_split(diabetes.data, diabetes.target, test_size=0.25, random_state=42)
```

```python
strategies = ['constant', 'quantile', 'mean', 'median']
baseline_args = [{'strategy': s} for s in strategies]
baseline_args[0]['constant']=50.
baseline_args[1]['quantile']=0.75
```

```python
def do_one(**args):
    baseline = dummy.DummyRegressor(**args)
    baseline.fit(train_data, train_target)
    basseline_pred = baseline.predict(test_data)
    return metrics.mean_squared_error(basseline_pred, test_target)
```

```python
mses = [do_one(**arg) for arg in baseline_args]
display(pd.DataFrame({'mse': mses, 'strategy': strategies}))
```
