Iz dataseta izbacite sve značajke koje nisu numeričke, a ako negdje nedostaje vrijednost umetnite medijan. 20% dataseta ostavite za testiranje, a pri dijeljenju dataset-a u funkciji train_test_split koristite `random_state=42` i `stratify=y`. Istrenirajte `k-NN` sa `5 susjeda` i ispišite koja je točnost modela.

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
diabetes_df = pd.read_csv("diabetes_dataset.csv", low_memory=False)
```

```python
diabetes_df.head()
```

```python
target=diabetes_df["Outcome"]
x_num = diabetes_df.select_dtypes(include=np.number)
imputer = SimpleImputer(strategy="median")
x_num_imputed = pd.DataFrame(imputer.fit_transform(x_num), columns=x_num.columns)
diabetes_df_cleaned = pd.concat([x_num_imputed, target], axis=1)
y = diabetes_df_cleaned["Outcome"].astype(int)
x = diabetes_df_cleaned.drop(columns=["Outcome"]).select_dtypes(include=np.number)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
algorithm = KNeighborsClassifier(n_neighbors=5)
model = algorithm.fit(x_train, y_train)
predictions = model.predict(x_test)
accuracy = accuracy_score(y_test, predictions)
```
