```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
from sklearn import model_selection
from sklearn import linear_model
```

```python
diabetes = datasets.load_diabetes()

linreg = linear_model.LinearRegression()
scores = []
loo = model_selection.LeaveOneOut()
scores = model_selection.cross_val_score(linreg, diabetes.data, diabetes.target, cv=loo, scoring='neg_mean_squared_error')
scores = pd.Series(np.sqrt(-scores))

df = pd.DataFrame({'RMSE': scores})
df.index.name = 'Repeat'
display(df.T)
ax = plt.figure(figsize=(4, 3)).gca()
sns.swarmplot(y='RMSE', data=df, ax=ax)
```
