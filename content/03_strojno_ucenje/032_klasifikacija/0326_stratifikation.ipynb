```python
from sklearn import datasets
from sklearn import neighbors
from sklearn import model_selection
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()

knn = neighbors.KNeighborsClassifier(10)
non_stratified_kf = model_selection.KFold(5)
izlaz = model_selection.cross_val_score(knn, iris.data, iris.target, cv=non_stratified_kf)
print(izlaz)

izlaz2 = model_selection.cross_val_score(knn, iris.data, iris.target, cv=5)
print(izlaz2)
```
