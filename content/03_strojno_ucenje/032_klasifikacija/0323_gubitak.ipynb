```python
import numpy as np
from sklearn import datasets, neighbors
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
def training_loss(loss, model, traingin_data):
    return sum(loss(model.predict(x.reshape(1, -1)), y) for x,y in traingin_data)

def squared_error(prediction, actual):
    return (prediction - actual) ** 2
```

```python
diabetes = datasets.load_diabetes()
knn = neighbors.KNeighborsRegressor()
model = knn.fit(diabetes.data, diabetes.target)
training_data = zip(diabetes.data, diabetes.target)
my_training_loss = training_loss(squared_error, model, training_data)
print(my_training_loss)
```

```python
# Ako koristimo mean_squared_error možemo to jednostavnije
```

```python
mse = mean_squared_error(diabetes.target, model.predict(diabetes.data))
print(mse*len(diabetes.data))
```
