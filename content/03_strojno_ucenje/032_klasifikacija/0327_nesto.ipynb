```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import datasets
from sklearn import model_selection
from sklearn import linear_model
from sklearn import metrics
```

```python
diabetes = datasets.load_diabetes()

linreg = linear_model.LinearRegression()
scores=[]
for i in range(10):
    (train_features, test_features, train_targets, test_targets) = model_selection.train_test_split(diabetes.data, diabetes.target, test_size=0.25)
    fit = linreg.fit(train_features, train_targets)
    predictions = fit.predict(test_features)
    score = metrics.mean_squared_error(test_targets, predictions)
    scores.append(score)
scores = pd.Series(np.sqrt(sorted(scores)))
df = pd.DataFrame({'RMSE': scores})
df.index.name = 'Repeat'
display(df.T)
ax = plt.figure(figsize=(4, 3)).gca()
sns.swarmplot(y='RMSE', x='Repeat', data=df, ax=ax)
```
