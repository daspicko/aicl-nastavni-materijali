Iz dataseta izbacite sve značajke koje nisu numeričke, a ako negdje nedostaje vrijednost umetnite medijan. Izbacite i target varijablu. Nakon toga standardizirajte podatke koristeći RobustScaler, pa istrenirajte hijerarhijski i DBSCAN algoritam za klasteriranje.


```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_csv("diabetes_dataset.csv", low_memory=False)
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
y = df["Outcome"]
X = df.drop(columns=["Outcome"])
```

```python
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

```python
hier_model = AgglomerativeClustering(n_clusters=2)
hier_labels = hier_model.fit_predict(X_scaled)
```

```python
dbscan_model = DBSCAN(eps=3, min_samples=5)
dbscan_labels = dbscan_model.fit_predict(X_scaled)
```

```python
ari_hier = adjusted_rand_score(y, hier_labels)
ami_hier = adjusted_mutual_info_score(y, hier_labels)

ari_db = adjusted_rand_score(y, dbscan_labels)
ami_db = adjusted_mutual_info_score(y, dbscan_labels)
```

```python
print("Hijerarhijsko klasteriranje")
print(f"Adjusted Rand Index (ARI): {ari_hier:.4f}")
print(f"Adjusted Mutual Information (AMI): {ami_hier:.4f}")

print("\nDBSCAN klasteriranje")
print(f"Adjusted Rand Index (ARI): {ari_db:.4f}")
print(f"Adjusted Mutual Information (AMI): {ami_db:.4f}")
```
