Koristite dataset [Breast Cancer Wisconsin (Diagnostic)](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data). Istrenirajte ansambl modele `AdaBoost` i `RandomForest`. Standardizacija u ovom slučaju nije potrebna jer ako se ne zada bazni model, koristi se stablo odlučivanja koje ne zahtijeva standardizaciju. Koristite `GridSearcCV`, a mijenjajte samo jedan hiperparametar po želji (dvije vrijednosti) i ispišite točnosti.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
```

U prethodnom zadatku smo ustanovili da je target varijabla "diagnosis" koja ima dvije vrijednosti: M (malignant) i B (benign). Dataset ima dodatni zarez na kraju svakog retka što uzrokuje stvaranje stupca "Unnamed: 32" s NaN vrijednostima. Taj stupac ćemo izbaciti kao i stupac ID.

```python
df = pd.read_csv("breast-cancer-wisconsin-data.csv", low_memory=False)
df = df.drop(columns=["Unnamed: 32", "id"])

target = df["diagnosis"].map({"M": 1, "B": 0})
features = df.drop(columns=["diagnosis"])
```

```python
ada = AdaBoostClassifier()

param_grid = {
    "n_estimators": [50, 100]
}
grid_ada = GridSearchCV(ada, param_grid, cv=5)
grid_ada.fit(features, target)

print("Best Estimator", grid_ada.best_estimator_,
      "Best Score:", grid_ada.best_score_,
      "Best Parameters:", grid_ada.best_params_,
      sep='\n')
```

```python
rf = RandomForestClassifier()
param_grid = {
    "n_estimators": [60, 120]
}
grid_rf = GridSearchCV(rf, param_grid, cv=5)
grid_rf.fit(features, target)

print("Best Estimator", grid_ada.best_estimator_,
      "Best Score:", grid_ada.best_score_,
      "Best Parameters:", grid_ada.best_params_,
      sep='\n')
```
