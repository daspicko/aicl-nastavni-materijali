```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
df = pd.read_csv("breast-cancer-wisconsin-data.csv", low_memory=False)
df = df.drop(columns=["Unnamed: 32", "id"])

target = df["diagnosis"].map({"M": 1, "B": 0})
features = df.drop(columns=["diagnosis"])
```

```python
algorithms = {
    "SVC_poly": SVC(kernel='poly'),
    "SVC_rbf": SVC(kernel='rbf')
}

param_grid = {
    "poly": {
        "degree": [2, 3]
    },
    "rbf": {
        "gamma": [0.1, 1.0]
    }
}

models = {
    "SVC_poly": GridSearchCV(algorithms["SVC_poly"], param_grid["poly"], cv=5).fit(features, target),
    "SVC_rbf": GridSearchCV(algorithms["SVC_rbf"], param_grid["rbf"], cv=5).fit(features, target),
}

for key in models.keys():
    print(f"Model: {key}")
    print("Best Estimator:", models[key].best_estimator_)
    print("Best Score:", models[key].best_score_)
    print("Best Parameters:", models[key].best_params_)
    print()

```
