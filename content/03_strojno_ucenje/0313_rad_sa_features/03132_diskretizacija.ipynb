```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import binarize
from sklearn.datasets import load_iris
```

### Diskretizacija

```python
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
display(iris_df.iloc[[0, 50, 100]])
```

```python
plt.subplots(1, 1, figsize=(4, 3))
ax = sns.distplot(iris_df['sepal length (cm)'], hist=False, rug=True)
ax.set_ylabel('Aproksimacija %')
```

```python
column = iris_df[['sepal length (cm)']]
column_mean = column.mean().values[0]
both = column.copy()
both['> Mean'] = binarize(column, threshold=column_mean).astype(bool)

display(both.iloc[[0, 50, 100]])
print('Mean:', column_mean)
```
