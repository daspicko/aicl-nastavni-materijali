# One Hot Encoder

[One Hot encoder](https://www.geeksforgeeks.org/machine-learning/ml-one-hot-encoding/)

```python
from sklearn import datasets
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
iris.feature_names = [fn[:-5] for fn in iris.feature_names]
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['spicies'] = iris.target_names[iris.target]
iris_df.head()
```

```python
print(iris.target[[0, 50, 100]])
```

```python
sparse = OneHotEncoder().fit_transform(iris.target.reshape(-1, 1))
print(sparse[[0, 1, 50, 51, 100, 101]].todense(), sep='\n')
```

One Hot Encoder metoda se može izvršiti i pomoću Pandasa. Detalji na [One Hot Encoding](https://www.geeksforgeeks.org/machine-learning/ml-one-hot-encoding/).

```python
encoded = pd.get_dummies(iris_df, prefix='is')
print(encoded.iloc[[0, 1, 50, 51, 100, 101]])
```

```python
X = encoded.drop(columns=['petal length'])
y = encoded['petal length']
```

```python
(features_train, features_test, target_train, target_test) = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
algorithm = LinearRegression()
model = algorithm.fit(features_train, target_train)
predictions = model.predict(features_test)

mse = mean_squared_error(target_test, predictions)
r2 = model.score(features_test, target_test)
print('MSE: %.3f' % mse)
print('R2: %.3f' % r2)
```

Izbacivanje columna

```python
df = pd.DataFrame({
    'grad': ['Zagreb', 'Split', 'Rijeka'],
    'cijena': [100, 120, 90]
})
X = df[['grad']]
y = df[['cijena']]
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False)
Z = ohe.fit_transform(X)
print(Z)
```

```python
lr = LinearRegression()
model_lr = lr.fit(Z, y)

print('Coef', model_lr.coef_)
print('Intercept', model_lr.intercept_)
```
