```python
import numpy as np
import pandas as pd
%matplotlib inline
from sklearn import neighbors
from sklearn.model_selection import GridSearchCV
from sklearn import datasets
```

```python
iris = datasets.load_iris()
diabetes = datasets.load_diabetes()
digits = datasets.load_digits()
```

```python
knn = neighbors.KNeighborsClassifier()
print(" ".join(knn.get_params().keys()))
```

```python
param_grid = {
    'n_neighbors': [1, 3, 5, 10, 20],
}
knn = neighbors.KNeighborsClassifier()
grid_model = GridSearchCV(knn, return_train_score=True, param_grid=param_grid, cv=10)
grid_model.fit(digits.data, digits.target)
print(grid_model)

parameters_columns = ['param_n_neighbors']
score_columns = ['mean_train_score', 'std_train_score', 'mean_test_score', 'std_test_score']

df = pd.DataFrame(grid_model.cv_results_).head()
```

```python
display(df[parameters_columns + score_columns])
```

```python
grid_df = pd.DataFrame(grid_model.cv_results_, columns=['param_n_neighbors', 'mean_train_score', 'mean_test_score'])
grid_df.set_index('param_n_neighbors', inplace=True)
display(grid_df)
```

```python
param_grid = {
    'n_neighbors': np.arange(1, 11),
    'weights': ['uniform', 'distance'],
    'p': [1, 2, 4, 8, 16]
}
knn = neighbors.KNeighborsClassifier()
grid_model = GridSearchCV(knn, return_train_score=True, param_grid=param_grid, cv=10)
%timeit -r1 grid_model.fit(iris.data, iris.target)
```

```python
param_df = pd.DataFrame.from_records(grid_model.cv_results_['params'])
param_df['mean_test_score'] = grid_model.cv_results_['mean_test_score']
param_df.sort_values(by=['mean_test_score']).tail()
```

```python
print("Best Estimator", grid_model.best_estimator_,
      "Best Score:", grid_model.best_score_,
      "Best Parameters:", grid_model.best_params_,
      sep='\n')
```
