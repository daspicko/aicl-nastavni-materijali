```python
def je_li_paran(n):
    return n % 2 == 0

dobitak = 0
for bacanje in range(10000):
    rezultat = np.random.randint(1, 7)
    dobitak += 1 if je_li_paran(rezultat) else -0.5

print(dobitak)
```
